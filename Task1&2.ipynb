```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def task1(time, start):
    dist = start
    dist_arr = []
    time_arr = [i for i in range(time)]

    for i in range(time):
        dist += np.random.choice([-1,0,1])
        dist_arr.append(dist)

    return (dist_arr, time_arr)

```

```python
dist_arr, time_arr = task1(1000, 0)
plt.plot( time_arr, dist_arr)
plt.show()
```
