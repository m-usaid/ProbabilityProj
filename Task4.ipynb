```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def task4(time, start):
    dist = start
    dist_arr = []
    time_arr = [i for i in range(time)]

    for i in range(time):
        dist += random.uniform(-1, 1)
        dist_arr.append(dist)

    return (dist_arr, time_arr)
```

```python
dist_arr, time_arr = task4(1000, 0)
plt.plot( time_arr, dist_arr)
plt.show()
```
