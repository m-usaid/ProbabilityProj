```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def task5(steps,x,y):
    theta_vals = [0]
    r_vals = [0, 0.5, 1]
    theta_val = 0
    x_vals, y_vals = [], []
    x,y = 0,0
        
    for i in range(steps):
        step = random.uniform(0,1)
        theta_step = random.uniform(0, 2*math.pi)
        x += step*math.cos(theta_step)
        y += step*math.sin(theta_step)
        x_vals.append(x)
        y_vals.append(y)
        if math.sqrt(x**2+y**2) > 100:
            x = -x
            y = -y
        
    return x_vals, y_vals
        
```

```python
def location_choice(num_loc):
    locs = []
    for i in range(num_loc):
        r = random.uniform(0,100)
        theta = random.uniform(0, 2*math.pi)
        x = round(r*math.cos(theta),2)
        y = round(r*math.sin(theta), 2)
        locs.append((x,y))
        
    return locs    
```

```python
locs = location_choice(2)
x_vals_a, y_vals_a = task5(1000, locs[0][0], locs[0][1])
x_vals_b, y_vals_b = task5(1000, locs[1][0], locs[1][1])
```

```python
plt.plot(x_vals_a, y_vals_a)
plt.plot(x_vals_b, y_vals_b)
plt.show()
```

**Simulation to find the Average Step**
-----------------------------------

```python
def RanSimulation(sim_num):
    locs = location_choice(2)
    x_valsa_mult = []
    y_valsa_mult = []
    x_valsb_mult = []
    y_valsb_mult = []
    for i in range(sim_num):
        x_vals_a, y_vals_a = task5(1000, locs[0][0], locs[0][1])
        x_vals_b, y_vals_b = task5(1000, locs[1][0], locs[1][1])
        x_valsa_mult.append(x_valsa)
        y_valsa_mult.append(y_valsa)
        x_valsb_mult.append(x_valsb)
        y_valsb_mult.append(y_valsb)
        
    return x_valsa_mult, y_valsa_mult, x_valsb_mult, y_valsb_mult
```
